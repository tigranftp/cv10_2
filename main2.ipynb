```python
idd = 0
table = []
for nameOfImage in listbasePhotoDIR: 
    print(idd, path)
    path = cwd + "\\" + basePhotoDIR + "\\" + nameOfImage
    image = preprocess(Image.open(path)).unsqueeze(0).to(device)   
    with torch.no_grad():
        vec = model.encode_image(image)[0].numpy()
    if len(vec) == 0:
        continue
    dicte = {"id": idd, "path":path, "vector":vec}
    idd+=1
    table.append(dicte)
```
